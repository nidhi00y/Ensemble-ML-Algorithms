```python
df.drop(['car_name','Unnamed: 0'],inplace=True,axis=1)
df.head()
```

```python
x = pd.DataFrame(df.drop('selling_price',axis=1))
y = df['selling_price']
x
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler

s = StandardScaler()

num = [f for f in x.columns if x[f].dtype != 'object']

x[num] = s.fit_transform(x[num])

encoders = {}

for col in x.columns:
    if x[col].dtype == 'object':
        le = LabelEncoder()
        x[col] = le.fit_transform(x[col])
        encoders[col] = le   

x.head(5)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=0000)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
r = GradientBoostingRegressor()
r.fit(x_train,y_train)
y_pred = r.predict(x_test)
```

```python
from sklearn.metrics import r2_score,root_mean_squared_error
print(r2_score(y_test,y_pred))
print(root_mean_squared_error(y_test,y_pred))
```
